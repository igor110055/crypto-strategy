## Get candle data

```python
import requests
import pandas as pd

token = 'bqqjd5frh5rcj5178ceg'

def get_symbol_list(exchange='binance', token=token):
    url = 'https://finnhub.io/api/v1/crypto/symbol?exchange={}&token={}'.format(exchange, token)
    r = requests.get(url)
    symbol_list = sorted([i['symbol'] for i in r.json()])
    return symbol_list
    
def get_table(symbol, start, end, resolution='D', token=token):
    url = 'https://finnhub.io/api/v1/crypto/candle?symbol={}&resolution={}&from={}&to={}&token={}'.format(symbol, resolution, start, end, token)
    r = requests.get(url)
    df = pd.DataFrame.from_dict(r.json())
    return df

symbol_list = get_symbol_list()
# print(symbol_list)
df = get_table('BINANCE:BTCUSDT', '1572651390', '1594080000')
df
```
